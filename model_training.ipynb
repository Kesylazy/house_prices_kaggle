```python
import pandas as pd
```

```python
train_data = pd.read_csv('train_upd.csv')
df_train = pd.DataFrame(train_data)
```

```python
test_data = pd.read_csv('test_upd.csv')
df_test = pd.DataFrame(test_data)
```

```python
import numpy as np
```

```python
X = df_train.drop('SalePrice', axis=1)
y = df_train['SalePrice']
y = np.log1p(y)
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=.2,
                                                    random_state=42)
```

```python
from sklearn.ensemble import StackingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import LinearRegression
```

```python
stack = StackingRegressor(
    estimators=[
        ('xgb', XGBRegressor()),
        ('lgb', LGBMRegressor()),
        ('cat', CatBoostRegressor(verbose=0))
    ],
    final_estimator=LinearRegression()
)
```

```python
stack.fit(X_train, y_train)
```

```python
y_pred = stack.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
y_pred_test = np.round(np.expm1(stack.predict(df_test)))
```

```python
y_pred_test
```

```python
submission_dictionary = {
    'Id': list(range(1461, 2920)),
    'SalePrice': y_pred_test
}
```

```python
submission_df = pd.DataFrame(submission_dictionary)
```
