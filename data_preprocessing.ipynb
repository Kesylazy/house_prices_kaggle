```python
# Delete the columns with more than 80% of nans
df = df.loc[:, df.isna().mean() < .8]
```

```python
# Delete the columns with more than 20% of nans
df = df.dropna(thresh=len(df.columns)/100*80) # PS, no column has been deleted
```

```python
# Create a df with the columns that contain the nans
df_nans = df.loc[:, df.isna().sum() != 0]
df_nans
```

# Code to fill/remove nans 

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
numerical_cols = df.select_dtypes(include=['number']).columns

for col in categorical_cols:
    mode_value = df[col].mode()[0]  # Calcola la moda (primo valore se ce ne sono più di uno)
    df[col].fillna(mode_value, inplace=True)

for col in numerical_cols:
    median_value = df[col].median()  # Calcola la mediana
    df[col].fillna(median_value, inplace=True)
```

```python
df.loc[1460:, 'SalePrice'] = 0
```

# Outliers management

```python
def identify_outliers(df, column):
    """
    Identify outliers in a column.
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```
